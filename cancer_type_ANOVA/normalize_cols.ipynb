```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import seaborn as sns
```

```python
brain_df = pd.read_csv('brain_gene_counts.txt', sep=' ', index_col='gene_id')
lung_df = pd.read_csv('lung_gene_counts.txt', sep=' ', index_col='gene_id')
panc_df = pd.read_csv('pancreas_gene_counts.txt', sep=' ', index_col='gene_id')
```

```python
# copy the data
brain_norm_df = brain_df.copy()

# apply normalization techniques
for column in brain_df.columns:
    brain_norm_df[column] = MinMaxScaler(feature_range=(0, 1)).fit_transform(np.array(brain_norm_df[column]).reshape(-1,1))

# view normalized data
display(brain_norm_df)
```

```python
lung_norm_df = lung_df.copy()

for column in lung_df.columns:
    lung_norm_df[column] = MinMaxScaler(feature_range=(0, 1)).fit_transform(np.array(lung_norm_df[column]).reshape(-1,1))

panc_norm_df = panc_df.copy()

for column in panc_df.columns:
    panc_norm_df[column] = MinMaxScaler(feature_range=(0, 1)).fit_transform(np.array(panc_norm_df[column]).reshape(-1,1))
```

# Below, I compare the distributions of one gene across cancer types

```python
brain_gene_x = list(brain_norm_df[brain_norm_df.index == 'ENSG00000000003.15'].values[0][:-1])
lung_gene_x = list(lung_norm_df[lung_norm_df.index == 'ENSG00000000003.15'].values[0][:-1])
panc_gene_x = list(panc_norm_df[panc_norm_df.index == 'ENSG00000000003.15'].values[0][:-1])
```

```python
sns.displot(brain_gene_x, kde=True)
```

```python
sns.displot(lung_gene_x, kde=True)
```

```python
sns.displot(panc_gene_x, kde=True)
```

```python
brain_norm_df = brain_norm_df.T
lung_norm_df = lung_norm_df.T
panc_norm_df = panc_norm_df.T
```

```python
brain_norm_df['tissue'] = 'brain'
lung_norm_df['tissue'] = 'lung'
panc_norm_df['tissue'] = 'pancreas'
```

```python
all_normalized = pd.concat([brain_norm_df, lung_norm_df, panc_norm_df])
```
